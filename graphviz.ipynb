```python
edges=sorted(G.edges(data=True), key=lambda t: t[0])
```

```python
import random
final_edges = []
for e in edges:
    fee = e[2]
    final_edges.append([e[0],e[1],fee])
    final_edges.append([e[1],e[0],fee])
ffe = sorted(final_edges, key=lambda t: t[0])
for e in ffe:
    print(str(e[0])+" "+str(e[1])+" "+str(e[2]['weight']))
```

```python
nx.shortest_path(G, "0", "3", weight='weight')
```

```python
import matplotlib.pyplot as plt
pos = nx.shell_layout(G)
nx.draw(G,pos)
# draw path in red
path = nx.shortest_path(G,source="0",target="3", weight='weight')
path_edges = list(zip(path,path[1:]))
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edges(G,pos,edgelist=path_edges,edge_color='r',width=1)
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.axis('equal')
plt.show()
```
